### Exercise 4-B: Euclidean norm (length)

**Goal:** Compute ‖x‖ = √(x₁² + x₂² + ... + xd²)

**Hint:**
```
Step 1: Square each element:  x ** 2
Step 2: Sum them:             np.sum(x**2)
Step 3: Take square root:     np.sqrt(...)
Shortcut:                     np.linalg.norm(x)
```

````python
x = np.array([1, 3, 4, 6, 8], dtype=float)
y = np.array([5, 1, 0, 2, 7], dtype=float)

# YOUR CODE

# Manual: sqrt( 1²+3²+4²+6²+8² )
norm_x_manual = np.sqrt(np.sum(x ** 2))
norm_x_numpy  = np.linalg.norm(x)

norm_y_manual = np.sqrt(np.sum(y ** 2))
norm_y_numpy  = np.linalg.norm(y)

print(f"‖x‖ manual: {norm_x_manual:.4f}")    # 11.2250
print(f"‖x‖ numpy:  {norm_x_numpy:.4f}")     # 11.2250
print(f"‖y‖ manual: {norm_y_manual:.4f}")    # 8.8882

# Normalise x (make its length = 1)
x_unit = x / norm_x_numpy
print(f"\n‖x_unit‖ = {np.linalg.norm(x_unit):.6f}")   # should be 1.0
````

### Exercise 4-C: Angle between vectors

**Goal:** Compute the angle θ between x and y.

**Hint:**
```
Formula: cos θ = xᵀy / (‖x‖ · ‖y‖)
Step 1: Compute dot product
Step 2: Compute both norms
Step 3: cos_theta = dot / (norm_x * norm_y)
Step 4: theta = np.arccos(cos_theta)   → radians
Step 5: np.degrees(theta)              → convert to degrees
```

````python
x = np.array([1, 3, 4, 6, 8], dtype=float)
y = np.array([5, 1, 0, 2, 7], dtype=float)

# YOUR CODE
dot_xy    = np.dot(x, y)                         # step 1
norm_x    = np.linalg.norm(x)                    # step 2a
norm_y    = np.linalg.norm(y)                    # step 2b
cos_theta = dot_xy / (norm_x * norm_y)           # step 3
theta_rad = np.arccos(cos_theta)                 # step 4
theta_deg = np.degrees(theta_rad)                # step 5

print(f"dot(x,y)  = {dot_xy}")
print(f"cos θ     = {cos_theta:.4f}")    # ~0.7613
print(f"θ (rad)   = {theta_rad:.4f}")    # ~0.7074
print(f"θ (deg)   = {theta_deg:.2f}°")   # ~40.5°

# Are they orthogonal? (dot = 0 means perpendicular)
print(f"\nOrthogonal? {np.isclose(dot_xy, 0)}")   # False
````

### Exercise 4-D: L1 vs L2 norm  (Lasso vs Ridge)

**Goal:** Understand why the choice of norm changes the penalty behavior.

**Hint:**
```
L2 norm:   ‖w‖  = sqrt(sum(wᵢ²))     → np.linalg.norm(w)
L2 norm²:  ‖w‖² = sum(wᵢ²)           → Ridge penalty
L1 norm:   ‖w‖₁ = sum(|wᵢ|)          → Lasso penalty
           np.sum(np.abs(w))
```

````python
# Two weight vectors: one has a near-zero value, one is spread out
w_sparse = np.array([2.0, 0.0, 0.0, 0.01, 0.0])   # mostly zeros
w_spread = np.array([0.4, 0.4, 0.4, 0.41, 0.4])   # spread across features

for name, w in [("w_sparse", w_sparse), ("w_spread", w_spread)]:
    l2_penalty = np.sum(w ** 2)            # Ridge: sum of squares
    l1_penalty = np.sum(np.abs(w))         # Lasso: sum of absolute values
    print(f"{name}: L2 penalty = {l2_penalty:.4f}  |  L1 penalty = {l1_penalty:.4f}")

print("""
Observation:
  L2 (Ridge): heavily penalises LARGE values, tolerates many small ones
  L1 (Lasso): penalises equally per unit → prefers fewer non-zero weights
  → L1 pushes weights exactly to 0 (sparse solution)
  → L2 shrinks weights but rarely to exactly 0
""")
````

## MODULE 5 — Determinant and Inverse

**Mental model:**
```
Determinant: one number that tells you if a matrix is invertible
  det ≠ 0  →  invertible  →  unique solution exists
  det = 0  →  singular    →  no solution (or infinite solutions)

Inverse M⁻¹: like dividing — M⁻¹ @ M = I

Linear Regression: W* = (XXᵀ)⁻¹ XY
  → needs (XXᵀ) to be invertible!
  Ridge fixes this: (XXᵀ + λI) is ALWAYS invertible
```

### Exercise 5-A: Compute and interpret the determinant

**Hint:**
```
np.linalg.det(M)   → determinant
det ≠ 0 → invertible
det = 0 → singular (no inverse)
```

````python
# Matrix from the lecture
A = np.array([[2,  4,  6],
              [-1, 2,  3],
              [1,  4,  9]], dtype=float)

# A singular matrix: row 2 = 2 × row 1 → redundant!
A_sing = np.array([[1, 2],
                   [2, 4]], dtype=float)

# YOUR CODE
det_A      = np.linalg.det(A)
det_A_sing = np.linalg.det(A_sing)

print(f"det(A)         = {det_A:.1f}")        # 24.0 → invertible
print(f"det(A_singular)= {det_A_sing:.6f}")   # 0.0  → not invertible

print(f"\nA invertible?       {not np.isclose(det_A,      0)}")  # True
print(f"A_sing invertible?  {not np.isclose(det_A_sing,  0)}")  # False

# Property: product of eigenvalues = determinant
eigs, _ = np.linalg.eig(A)
print(f"\nProduct of eigenvalues: {np.prod(eigs).real:.2f}")   # should match det
print(f"Determinant:            {det_A:.2f}")
````

````python
# ✅ Check
check("det(A)",       det_A,      24.0)
check("det(A_sing)",  det_A_sing, 0.0, tol=1e-9)
````

**Hint:**

### Exercise 5-C: Why Ridge always has a solution

**Goal:** Show that adding λI to XXᵀ makes it always invertible.

**Hint:**
```
Step 1: Create X where the rows are linearly dependent
        → XXᵀ will be singular (det=0)
Step 2: Add λI  → (XXᵀ + λI) → det > 0 → invertible
Step 3: Print eigenvalues to see why:
        eigenvalues of XXᵀ  include 0
        eigenvalues of (XXᵀ + λI)  = previous + λ  (all > 0)
```

````python
# X where row 2 = 2 × row 1 (linearly dependent → rank 1)
X = np.array([[1, 2, 3],
              [2, 4, 6]], dtype=float)

XXT = X @ X.T                  # shape (2, 2)
lam = 1.0

# YOUR CODE
XXT_ridge = XXT + lam * np.eye(XXT.shape[0])

det_XXT       = np.linalg.det(XXT)
det_XXT_ridge = np.linalg.det(XXT_ridge)

print(f"det(XXT)          = {det_XXT:.4f}")       # 0.0 → singular!
print(f"det(XXT + λI)     = {det_XXT_ridge:.4f}")  # > 0  → invertible

eigs_before = np.linalg.eigvalsh(XXT)
eigs_after  = np.linalg.eigvalsh(XXT_ridge)
print(f"\nEigenvalues of XXT      : {np.round(eigs_before, 4)}")   # includes 0
print(f"Eigenvalues of XXT+λI   : {np.round(eigs_after,  4)}")    # all > 0
print("\nConclusion: adding λI shifts every eigenvalue by λ → no zeros → always invertible")
````

## MODULE 6 — Eigenvalues & Eigenvectors

**Mental model:**
```
For matrix M:  M @ u = λ @ u

u = eigenvector:  a special direction that M doesn't rotate
λ = eigenvalue:   how much M stretches along that direction

In PCA: eigenvectors = principal component directions
        eigenvalues  = how much variance in each direction
```

### Exercise 6-A: Compute and verify eigenpairs

**Hint:**
```
Step 1: np.linalg.eig(M)   → returns (eigenvalues, eigenvectors)
        eigenvectors are columns of the returned matrix
Step 2: Extract i-th eigenvector: vecs[:, i]
Step 3: Verify: M @ u  should equal  λ * u
        np.allclose(M @ u, lam * u)  should be True
```

````python
M = np.array([[3.0, 1.0],
              [0.0, 2.0]])

# YOUR CODE
eigenvalues, eigenvectors = np.linalg.eig(M)

print("Eigenvalues:",  np.round(eigenvalues, 4))     # [3.0, 2.0]
print("Eigenvectors (columns):\n", np.round(eigenvectors, 4))

# Verify M @ u = λ * u for each pair
for i in range(len(eigenvalues)):
    lam = eigenvalues[i]
    u   = eigenvectors[:, i]    # i-th column
    Mu  = M @ u
    lu  = lam * u
    ok  = np.allclose(Mu, lu)
    print(f"\nPair {i+1}: λ={lam:.1f}")
    print(f"  M @ u  = {np.round(Mu, 4)}")
    print(f"  λ × u  = {np.round(lu, 4)}")
    print(f"  Match? {ok}")
````

### Exercise 6-B: Properties — sum, product, trace

**Hint:**
```
Property 1: product of eigenvalues = determinant
Property 2: sum of eigenvalues = trace
Property 3: all eigenvalues > 0 → matrix is PD → gradient descent converges
```

````python
M = np.array([[4.0, 2.0],
              [1.0, 3.0]])

# YOUR CODE
eigs, _ = np.linalg.eig(M)

# Property 1: product of eigenvalues = det
prod_eigs = np.prod(eigs)
det_M     = np.linalg.det(M)
print(f"Product of eigenvalues: {prod_eigs.real:.4f}")
print(f"Determinant:            {det_M:.4f}")
print(f"Match? {np.isclose(prod_eigs.real, det_M)}")

# Property 2: sum of eigenvalues = trace
sum_eigs = np.sum(eigs)
trace_M  = np.trace(M)
print(f"\nSum of eigenvalues: {sum_eigs.real:.4f}")
print(f"Trace:              {trace_M:.4f}")
print(f"Match? {np.isclose(sum_eigs.real, trace_M)}")

# Property 3: PD check
print(f"\nAll eigenvalues > 0? {np.all(eigs.real > 0)}  → M is PD")
````

### Exercise 6-C: Eigendecomposition  M = UΛUᵀ

**Hint:**
```
Step 1: np.linalg.eigh(M)  → for symmetric M, returns sorted real eigenvalues
Step 2: np.diag(eigs)      → build diagonal Λ from eigenvalues
Step 3: U @ Lambda @ U.T   → reconstruct M
Step 4: np.allclose(M_reconstructed, M)  → should be True
```

````python
M_sym = np.array([[4.0, 2.0],
                  [2.0, 3.0]])   # symmetric: M == Mᵀ

# YOUR CODE
eigs, U = np.linalg.eigh(M_sym)    # eigh for symmetric
Lambda  = np.diag(eigs)             # diagonal matrix of eigenvalues

M_reconstructed = U @ Lambda @ U.T  # M = U Λ Uᵀ

print("Eigenvalues:", np.round(eigs, 4))
show("U (eigenvectors)", U)
show("Lambda", Lambda)
show("Reconstructed M = UΛUᵀ", M_reconstructed)

print("\nM == UΛUᵀ?",      np.allclose(M_sym, M_reconstructed))  # True
print("U is orthogonal?",  np.allclose(U @ U.T, np.eye(2)))       # True
````

## MODULE 7 — PCA

**Mental model:**
```
Problem: 4 correlated features → hard to visualise, redundant info
PCA solution: find the 2 directions of MOST VARIATION
              project data onto those 2 directions
              keep 92% of information with only 50% of features

The directions = eigenvectors of the covariance matrix
The variation  = eigenvalues (bigger = more important)
```

### Exercise 7-A: Variance explained

**Hint:**
```
Step 1: total variance = sum of ALL eigenvalues
Step 2: variance from k components = sum of TOP k eigenvalues
Step 3: % = (step 2 / step 1) × 100
```

````python
# Eigenvalues from the lecture example
eigenvalues = np.array([16.5, 5.4, 1.5, 0.4])   # sorted largest first

# YOUR CODE
total_var   = np.sum(eigenvalues)
var_2_PCs   = np.sum(eigenvalues[:2])    # top 2
pct_2_PCs   = var_2_PCs / total_var * 100

print(f"Total variance:         {total_var}")
print(f"Variance from 2 PCs:    {var_2_PCs}")
print(f"% explained by 2 PCs:   {pct_2_PCs:.2f}%")   # 92.02%

# Show cumulative for all components
print("\nCumulative variance explained:")
for k in range(1, 5):
    cum = np.sum(eigenvalues[:k]) / total_var * 100
    bar = '█' * int(cum // 5)
    print(f"  PC{k}: {cum:5.1f}%  {bar}")
````

````python
# ✅ Check
check("% explained (2 PCs)", pct_2_PCs, 92.02, tol=0.1)
````

### Exercise 7-C: Full PCA pipeline from scratch

**Hint:**
```
Step 1: Mean-centre data:       X_c = X - X.mean(axis=0)
Step 2: Covariance matrix:      C = (X_c.T @ X_c) / n_samples
Step 3: Eigendecomposition:     np.linalg.eigh(C)  → sorted ascending
Step 4: Reverse order:          eigs[::-1],  vecs[:, ::-1]
Step 5: Take top k eigenvectors: Uk = vecs[:, :k]
Step 6: Project:                data_2d = X_c @ Uk
```

````python
np.random.seed(42)
n = 100

# Simulated body data: height, weight, neck, waist, age
height = np.random.normal(1.70, 0.10, n)
weight = 40 * height + np.random.normal(0, 5, n)      # correlated with height
neck   = 0.2 * weight + np.random.normal(0, 2, n)     # correlated with weight
waist  = 0.5 * weight + np.random.normal(0, 5, n)     # correlated with weight
age    = np.random.normal(40, 10, n)                   # independent

X = np.column_stack([height, weight, neck, waist, age])  # shape (100, 5)
print("Data shape:", X.shape)    # (100, 5)

# YOUR CODE — implement each step

# Step 1: Mean-centre
X_c = X - X.mean(axis=0)

# Step 2: Covariance matrix
C = (X_c.T @ X_c) / n
print("Covariance matrix shape:", C.shape)    # (5, 5)

# Step 3: Eigendecomposition
eigs, vecs = np.linalg.eigh(C)

# Step 4: Sort descending
eigs = eigs[::-1]
vecs = vecs[:, ::-1]

# Step 5: Top 2 eigenvectors
k  = 2
Uk = vecs[:, :k]         # shape (5, 2)

# Step 6: Project
data_2d = X_c @ Uk        # shape (100, 2)

print("\nTop 2 eigenvalues:",   np.round(eigs[:2], 2))
print("Variance explained: ",  round(np.sum(eigs[:2]) / np.sum(eigs) * 100, 1), "%")
print("Projected shape:",       data_2d.shape)   # (100, 2)

# Verify with sklearn
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

X_std   = StandardScaler().fit_transform(X)
pca_sk  = PCA(n_components=2)
data_sk = pca_sk.fit_transform(X_std)

print("\nsklearn explained variance:",
      np.round(pca_sk.explained_variance_ratio_ * 100, 1), "%")
````

## MODULE 8 — Linear Regression: Full Pipeline

**This brings everything together:**
```
1. Data → vector/matrix form
2. Standardise → norms (Module 4)

4. Predict → matrix multiply (Module 3)
5. Evaluate → norms, MSE
6. Ridge → eigenvalues ensure invertibility (Module 6)
```

### Exercise 8-A: Linear Regression from scratch

**Hint:**
```
Step 1: Standardise X:  mean = X.mean(axis=1),  std = X.std(axis=1)
                        X_sc = (X - mean[:,None]) / std[:,None]
        [:,None] reshapes (2,) to (2,1) for broadcasting

Step 2: Add bias row:   np.vstack([X_sc, np.ones((1, m))])

Step 3: Solve:          XXT = X @ X.T
                        XY  = X @ Y
                        W   = np.linalg.inv(XXT) @ XY

Step 4: Predict:        Y_pred = X.T @ W

Step 5: MSE:            np.mean((Y_pred - Y) ** 2)
```

````python
# Dataset: 4 houses  [dien_tich, so_phong]  →  gia (ty dong)
# Design matrix format: each COLUMN = one house
X_raw = np.array([[50, 80, 60, 90],    # x1: dien_tich
                  [ 2,  3,  2,  4]],   # x2: so_phong
                 dtype=float)          # shape (2, 4)

Y = np.array([[3.0],
              [4.5],
              [3.5],
              [5.5]])                  # shape (4, 1)

# YOUR CODE

# Step 1: Standardise each feature row
mean_x = X_raw.mean(axis=1)           # mean per feature → shape (2,)
std_x  = X_raw.std(axis=1)            # std  per feature → shape (2,)
X_sc   = (X_raw - mean_x[:, None]) / std_x[:, None]

print("mean:", np.round(mean_x, 4))   # [70.  2.75]
print("std: ", np.round(std_x,  4))   # [15.8114  0.8292]

# Step 2: Append bias row of 1s
X = np.vstack([X_sc, np.ones((1, X_sc.shape[1]))])   # shape (3, 4)
print("\nX (with bias):\n", np.round(X, 4))

# Step 3: Solve W* = (XXT)⁻¹ XY
XXT  = X @ X.T
XY   = X @ Y
W    = np.linalg.inv(XXT) @ XY

print("\nW* = [w1, w2, bias]ᵀ:")
print(np.round(W, 4))        # w1≈0.5534, w2≈0.4146, bias≈4.125

# Step 4: Predict
Y_pred = X.T @ W
print("\nPredicted prices:", np.round(Y_pred.flatten(), 4))
print("Actual prices:   ", Y.flatten())

# Step 5: MSE
mse = np.mean((Y_pred - Y) ** 2)
print(f"\nMSE = {mse:.6f}")   # 0.00625
print(f"RMSE = {np.sqrt(mse):.4f} ty dong ≈ {np.sqrt(mse)*1000:.0f} trieu")
````

````python
# ✅ Check
check("w1",   float(W[0]), 0.5534, tol=0.01)
check("w2",   float(W[1]), 0.4146, tol=0.01)
check("bias", float(W[2]), 4.125,  tol=0.01)
check("MSE",  mse,          0.00625)
````

### Exercise 8-B: Ridge Regression

**Hint:**
```
Only one change from Linear Regression:
  W = inv(XXT + λI) @ XY
  instead of
  W = inv(XXT) @ XY

Try different values of λ and see how W changes.
```

````python
# Reuse X, Y, XXT, XY from Exercise 8-A

print(f"{'λ':>6}  {'w1':>8}  {'w2':>8}  {'bias':>8}  {'MSE':>10}")
print("-" * 50)

for lam in [0.0, 0.1, 1.0, 5.0, 20.0]:
    # YOUR CODE: only change from linear is adding lam * I
    XXT_r  = XXT + lam * np.eye(XXT.shape[0])   # add λI
    W_r    = np.linalg.inv(XXT_r) @ XY
    Y_r    = X.T @ W_r
    mse_r  = float(np.mean((Y_r - Y)**2))
    print(f"{lam:>6.1f}  {float(W_r[0]):>8.4f}  {float(W_r[1]):>8.4f}  {float(W_r[2]):>8.4f}  {mse_r:>10.6f}")

print("""
Observations:
  λ=0.0 → same as Linear Regression
  λ↑    → w1, w2 shrink toward 0 (but never exactly 0)
  λ↑    → bias stays close to mean(Y) = 4.125
  λ↑    → train MSE increases (accepting more bias)
  λ↑    → model becomes more stable (less variance)
""")
````

|---|---|

| Stack cols | `np.hstack([a, b])` |

| Identity n×n | `np.eye(n)` |
| Diagonal matrix | `np.diag([a,b,c])` |
| Determinant | `np.linalg.det(M)` |
| Inverse | `np.linalg.inv(M)` |
| Eigenvalues | `np.linalg.eig(M)` |
| Eigenvalues (symmetric) | `np.linalg.eigh(M)` |
| Trace | `np.trace(M)` |
